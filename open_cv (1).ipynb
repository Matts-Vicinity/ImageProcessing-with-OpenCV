# Resizing and Rescaling videos and images

```python
import cv2 as cv

def rescale(frame, scale = 0.10):
    width = int(frame.shape[1] * scale)
    height = int(frame.shape[0] * scale)

    dimensions = (width,height)
    return cv.resize(frame, dimensions, interpolation = None)

video_capture = cv.VideoCapture(0)
while True:    
    isTrue, frame = video_capture.read()
    frame_resized = rescale(frame)
    cv.imshow('video', frame)
    cv.imshow('video_resized', frame_resized)
    if cv.waitKey(10) & 0xFF == ord('a'):
        break

video_capture.release()
cv.destroyAllWindows()
```

# Drawing and Writing on Images

```python
import cv2 as cv
import numpy as np

blank = np.zeros((500, 500, 3), dtype = 'uint8')

#pointing the image a certain color
blank[250:250, 250:250] = 255,255,255
cv.imshow('blue', blank)

#drawing a rectangle
cv.rectangle(blank, (0,0), (250,250), (255,0,0), thickness = 2)
cv.imshow('rectangle', blank)
cv.waitKey(0)
cv.destroyAllWindows

#drawing a circle
cv.circle(blank, (blank.shape[1]//2,blank.shape[0]//2), 100, (0,255,0), thickness = 3)
cv.imshow('circle',blank)
cv.waitKey(0)
cv.destroyAllWindows

#drawing a circle
cv.circle(blank, (blank.shape[1]//2,blank.shape[0]//2), 100, (0,255,0), thickness = 3)
cv.imshow('line',blank)
cv.waitKey(0)
cv.destroyAllWindows

#writing a text
cv.putText(blank, 'Hello Abhishek', (250,250), cv.FONT_HERSHEY_TRIPLEX, 2, (0,255,255), 2)
cv.imshow('text', blank)
cv.waitKey(0)
cv.destroyAllWindows
```

```python
#converting an image to grey scale image
import cv2 as cv
import numpy as np

def resize(image1, scale = 0.1):
    width = int(image1.shape[1]*scale)
    height = int(image1.shape[0]*scale)
    dimensions = (width,height)
    return cv.resize(image1, dimensions, interpolation = None)
                     
image1 = cv.imread('image.JPG')
resizeimage = resize(image1)

#converting into grey image
gray = cv.cvtColor(resizeimage, cv.COLOR_BGR2GRAY)
cv.imshow('resized_image', gray)

#blurring an image
blur = cv.GaussianBlur(resizeimage, (5,5), cv.BORDER_DEFAULT)
cv.imshow('blurred_image', blur)

#edge cascade
canny = cv.Canny(blur, 120, 120)
cv.imshow('edged_image', canny)
cv.imshow('original_image',resizeimage)

#dialating an image
dilate = cv.dilate(canny, (7,7), iterations = 4)
cv.imshow('dilation', dilate)

#eroding an image
erode = cv.erode(dilate, (3,3), iterations = 1)
cv.waitKey(0)
```

# color spaces using opencv

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

image = cv.imread('image.JPG')
resize = cv.resize(image, (200,200), interpolation = cv.INTER_CUBIC)

#BGR TO GREY
gray = cv.cvtColor(resize,cv.COLOR_BGR2GRAY)
cv.imshow('image',resize)
cv.imshow('image',gray)

#BGR TO HSV(HUGE SPACE VALUE)
hsv = cv.cvtColor(resize, cv.COLOR_BGR2HSV)
cv.imshow('hsv_image',hsv)

#BGR TO LAB
lab = cv.cvtColor(resize,cv.COLOR_BGR2LAB)
cv.imshow('lab_image',lab)

#BGR TO RGB
rgb = cv.cvtColor(resize,cv.COLOR_BGR2RGB)
cv.imshow('rgb_image',rgb)

plt.imshow(rgb)
plt.show()
cv.waitKey(0)
```
